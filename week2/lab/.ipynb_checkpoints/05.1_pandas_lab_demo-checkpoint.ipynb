### 8.  Which continent has the highest on average wine consumption?

```python
drinks[['wine_servings','continent']].groupby('continent').wine_servings.mean().sort_values(ascending=False).head(5)
```

```python
item_iters = drinks.groupby('continent').wine_servings.mean().sort_values(ascending=False).iteritems()
print item_iters.next()
print item_iters.next()
```
